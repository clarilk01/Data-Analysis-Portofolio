```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('..data/equipment_anomaly_data.csv')
```

```python
df.head()
```

```python
mask = df['equipment'] == "Pump"
filtered_df = df[mask]        
```

```python
filtered_df.head()
```

```python
def max_pressure_anomaly(dataframe):
    max_pressure = dataframe['pressure'].max()
    return max_pressure
result = max_pressure_anomaly(filtered_df)
print("Maximum Pressure Anomaly for Pump Equipment:", result)
```
